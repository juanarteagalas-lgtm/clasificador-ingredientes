# Clasificador de Ingredientes y Calorías 

### Proyecto de Inteligencia Artificial
Este proyecto reconoce ingredientes en una imagen y muestra
su información nutricional y calorías.

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt
import numpy as np
import os

# Parámetros
img_size = (224, 224)
batch_size = 8

# Generadores de imágenes
train_gen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    zoom_range=0.2,
    horizontal_flip=True
)

test_gen = ImageDataGenerator(
    rescale=1./255
)

# Cargar datos de entrenamiento
train_data = train_gen.flow_from_directory(
    directory='data/train',
    target_size=img_size,
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=True
)

# Cargar datos de prueba
test_data = test_gen.flow_from_directory(
    directory='data/test',
    target_size=img_size,
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=False
)

# Eliminar manualmente clases no deseadas (como carpetas ocultas)
valid_classes = ['azucar', 'harina', 'huevos', 'mantequilla', 'manzanas']

train_data.class_indices = {
    k: v for k, v in train_data.class_indices.items() if k in valid_classes
}

print("Clases finales usadas por el modelo:")
print(train_data.class_indices)

```

```python
model = models.Sequential()

model.add(layers.Conv2D(32, (3,3), activation='relu', input_shape=(224,224,3)))
model.add(layers.MaxPooling2D(2,2))

model.add(layers.Conv2D(64, (3,3), activation='relu'))
model.add(layers.MaxPooling2D(2,2))

model.add(layers.Conv2D(128, (3,3), activation='relu'))
model.add(layers.MaxPooling2D(2,2))

model.add(layers.Flatten())
model.add(layers.Dense(128, activation='relu'))
model.add(layers.Dense(5, activation='softmax'))

model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

model.summary()
```

```python
history = model.fit(
    train_data,
    epochs=10,
    validation_data=test_data
)
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import layers, models
```

```python
img_size = (224, 224)
batch_size = 8

train_gen = ImageDataGenerator(rescale=1./255)
test_gen = ImageDataGenerator(rescale=1./255)

train_data = train_gen.flow_from_directory(
    'data/train',
    target_size=img_size,
    batch_size=batch_size,
    class_mode='categorical'
)

test_data = test_gen.flow_from_directory(
    'data/test',
    target_size=img_size,
    batch_size=batch_size,
    class_mode='categorical'
)
```

```python
model = models.Sequential([
    layers.Input(shape=(224,224,3)),

    layers.Conv2D(32, (3,3), activation='relu'),
    layers.MaxPooling2D(2,2),

    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D(2,2),

    layers.Conv2D(128, (3,3), activation='relu'),
    layers.MaxPooling2D(2,2),

    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(5, activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

model.summary()
```

```python
history = model.fit(
    train_data,
    epochs=10,
    validation_data=test_data
)
```

```python
model.save("modelo_ingredientes.h5")
print("Modelo guardado correctamente")
```

```python
import numpy as np
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
from PIL import Image
import matplotlib.pyplot as plt

# ---------------------------
# 1. Configuración
# ---------------------------
IMG_SIZE = (224, 224)
MODEL_PATH = "modelo_ingredientes.h5"

# Clases del modelo (orden fijo)
clases = ['azucar', 'harina', 'huevos', 'mantequilla', 'manzanas']

# ---------------------------
# 2. Cargar modelo
# ---------------------------
model = load_model(MODEL_PATH)

# ---------------------------
# 3. Ruta de la imagen a probar
# ---------------------------
img_path = "data/prueba/manzana_test.jpeg"  # CAMBIA SI ES NECESARIO

# ---------------------------
# 4. Cargar y preparar imagen
# ---------------------------
img = image.load_img(img_path, target_size=IMG_SIZE)
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0)
img_array = img_array / 255.0

# ---------------------------
# 5. Predicción
# ---------------------------
predicciones = model.predict(img_array)

indice = np.argmax(predicciones)
confianza = np.max(predicciones) * 100

# ---------------------------
# 6. Resultado
# ---------------------------
print(f"Ingrediente detectado: {clases[indice]}")
print(f"Confianza: {confianza:.2f}%")

# Mostrar imagen
plt.imshow(img)
plt.axis("off")
```
